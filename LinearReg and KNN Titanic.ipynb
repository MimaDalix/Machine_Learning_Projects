```python
#regression avec SKlearn

np.random.seed(0)
m = 100  #create 100 samples
X = np.linspace(0, 10, m).reshape(m, 1)
y = X + np.random.randn(m, 1)
```

```python
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X, y)
model.score(X, y)

predictions = model.predict(X)

plt.scatter(X, y)
plt.plot(X, predictions, c='r')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from  sklearn.datasets import load_iris
```

```python
iris = load_iris()

X = iris.data
y = iris.target

print(X.shape)
plt.scatter(X[:, 0], X[:,1], c=y, alpha=0.8)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 5)

print('train test:', X_train.shape)
print('test test:', X_test.shape)
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.scatter(X_train[:, 0], X_train[:,1], c=y_train, alpha=0.8)
plt.title("Train set")
plt.subplot(122)
plt.scatter(X_test[:, 0], X_test[:,1], c=y_test, alpha=0.8)
plt.title("Test set")
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=6)

model.fit(X_train, y_train)
print('Train score: ', model.score(X_train, y_train))
print('Test score: ', model.score(X_test, y_test))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(KNeighborsClassifier(5), X_train, y_train, cv=5, scoring='accuracy').mean()
```

```python
val_score = []
for k in range(1, 50):
    score = cross_val_score(KNeighborsClassifier(k), X_train, y_train, cv=5, scoring='accuracy').mean()
    val_score.append(score)
    
plt.plot(val_score)
    
```

```python
from sklearn.model_selection import validation_curve
```

```python
model = KNeighborsClassifier()
k = np.arange(1, 50)

train_score, val_score = validation_curve(model, X_train, y_train, 'n_neighbors', k, cv=5)

plt.plot(k, val_score.mean(axis=1))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_neighbors': np.arange(1, 20), 'metric':['euclidian', 'manhattan']}

grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
grid.fit(X_train, y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
model = grid.best_estimator_
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, model.predict(X_test))
```

```python
from sklearn.model_selection import learning_curve
```

```python
N, train_score, val_score = learning_curve(model, X_train, y_train, train_sizes= np.linspace(0.1, 1.0, 10), cv=5)

print(N)
plt.plot(N, train_score.mean(axis=1), label='train')
plt.plot(N, val_score.mean(axis=1), label='validation')
plt.xlabel('Train_sizes')
plt.legend()
```
