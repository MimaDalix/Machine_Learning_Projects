```python
import numpy as np

```

```python
x = np.linspace(0, 2, 10)
y = x**2
print(x)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.plot(x, y, ls='--', label='quadratique')
plt.plot(x, x**3, label='cubique')
plt.title('figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.show()
plt.savefig('figure1.png')

plt.figure()
plt.plot(x, y)
```

```python
plt.scatter(x, y, c='pink', lw=5)
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y, c='red')
plt.title('Graphique 1')
plt.subplot(2, 1, 2)
plt.plot(x, y, c='purple')
```
